```python
from langchain_groq import ChatGroq
from langchain_core.messages import HumanMessage


llm = ChatGroq(
    model_name="llama3-8b-8192",  
    groq_api_key="API_key"  # Replace with your actual API key
)

# Few-shot prompt with examples
prompt = """You are a helpful assistant that rates semantic similarity between sentence pairs.
Rate similarity on a scale from 0 (completely different) to 5 (identical meaning).

Examples:

1. "A man is playing guitar."  
   "Someone is performing music."  
   → Similarity Score: 4.5

2. "A child is eating ice cream."  
   "An adult is reading a book."  
   → Similarity Score: 0.5

3. "A dog is running in the park."  
   "An animal is sprinting in a grassy area."  
   → Similarity Score: 4.8

Now rate the following:

4. "She is typing on her laptop."  
   "A woman is writing on her computer."  
   → Similarity Score:"""

# Send prompt as a message
response = llm.invoke([HumanMessage(content=prompt)])

# Print the response content
print("Model output:\n", response.content)
```
